**OUTLIER DETECTION USING IQR METHOD**

**Calculating lower and upper bounds for each numerical feature**

```python
# INTER-QUARTILE RANGE
for col in ['Temperature Min', 'Temperature Max', 'pH Min', 'pH Max', 'Conductivity (µmhos/cm) Min', 'Conductivity (µmhos/cm) Max']:

    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)

    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[
        (df[col] < lower_bound) |
        (df[col] > upper_bound)
    ]

    print(f"\n{col}")
    print(f"Lower Bound: {lower_bound:.2f}")
    print(f"Upper Bound: {upper_bound:.2f}")
    print(f"Outliers: {len(outliers)}")
```

**SCORES FOR EVERY COLUMN FUNCTIONS**

```python
def ph_Score(ph):
    if(ph>=6.5 and ph<=8.5):
        return 100
    elif((ph>=6.0 and ph<=6.5)or (ph>=8.5 and ph<=9.0)):
        return 70
    else:
        return 40
df_fe['pH_Score'] = df_fe['pH_Avg'].apply(ph_Score)
```

```python
def cond_score(cond):
    if cond <= 5.76:
        return 100
    elif cond <= 7.30:
        return 70
    else:
        return 40

df_fe['Cond_Score'] = df_fe['Conductivity_Log_Avg'].apply(cond_score)
```

```python
df_fe['WQI'] = ( df_fe['pH_Score'] +df_fe['Cond_Score']) / 2
```

```python
def classify_wqi(wqi):

    if wqi >= 90:
        return 'Excellent'

    elif wqi >= 75:
        return 'Good'

    else:
        return 'Poor'

df_fe['Water_Quality'] = df_fe['WQI'].apply(classify_wqi)
```

**LABEL ENCODING**   

The target variable Water_Quality contains categorical values (Excellent, Good, and Poor). Since machine learning algorithms require numerical input, Label Encoding is applied to convert these categories into integer labels.

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(df_fe['Water_Quality'])
```

```python
le.classes_
print(le.classes_)
```

**Train-Test Split**

*The dataset is divided into training and testing sets using the train_test_split() function.*

*Training Data (80%): Used to train the Logistic Regression model.*

*Testing Data (20%): Used to evaluate the model on unseen data.*

*The parameter random_state = 42 ensures that the same split is generated every time the code is executed, making the results reproducible.*

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y,test_size=0.2,random_state=42)
```

**Logistic Regression Model Training**

*Logistic Regression is a supervised machine learning algorithm used for classification tasks. In this project, it is used to classify water quality into the categories Excellent, Good, and Poor.*

*The model is initialized with max_iter = 1000, which allows the optimization algorithm sufficient iterations to converge to the best solution.*

*The fit() method trains the model using the training dataset. During training, the algorithm learns the relationship between the input water quality parameters and the corresponding water quality classes.*

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=1000)
model.fit(x_train, y_train)
```

**Model Prediction**

*After training, the Logistic Regression model is used to predict the water quality class for each sample in the testing dataset.*

*The predicted class labels are stored in "y_pred" and are later compared with the actual test labels to evaluate model performance.*

```python
y_pred = model.predict(x_test)
```

**Confusion Matrix**

*A Confusion Matrix is used to evaluate the classification performance of the model.*

*Rows represent the actual classes.*
*Columns represent the predicted classes.*

*The diagonal values represent correctly classified samples, while the off-diagonal values represent misclassifications.*

*This matrix helps identify which classes are being confused by the model.*

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test, y_pred))
```

**Classification Report**

*The Classification Report provides detailed performance metrics for each class.*

*Metrics included:*

*Precision: Measures how many predicted samples of a class are actually correct.*

*Recall: Measures how many actual samples of a class are correctly identified.*

*F1-Score: Harmonic mean of Precision and Recall.*

*Support: Number of actual samples belonging to each class.*

*These metrics provide a more detailed evaluation than accuracy alone.*

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

**Model Accuracy**

*Accuracy measures the overall correctness of the model's predictions.*

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
